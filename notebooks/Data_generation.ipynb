```python
import pandas as pd
import numpy as np
from faker import Faker

fake = Faker()

np.random.seed(42)
```

```python
# Generate Customer Dataset
customers=[]

for i in range(1,1001):

    customers.append({

        "customer_id":i,

        "age":np.random.randint(18,60),

        "monthly_income":np.random.randint(10000,150000),

        "occupation":np.random.choice([
            "Student",
            "Teacher",
            "Engineer",
            "Driver",
            "Shop Owner",
            "Freelancer"
        ]),

        "city":fake.city(),

        "state":np.random.choice([
            "Maharashtra",
            "Gujarat",
            "Karnataka",
            "Delhi",
            "Telangana"
        ])
    })

customer_df=pd.DataFrame(customers)

customer_df.head()
```

```python
# save customer data
customer_df.to_csv(
    "../data/customers.csv",
    index=False
)

print("customers.csv saved")
```

```python
# Generate UPI Dataset
upi=[]

for customer in customer_df["customer_id"]:

    upi.append({

        "customer_id":customer,

        "txn_count":
        np.random.randint(20,300),

        "avg_transaction":
        np.random.randint(50,5000),

        "grocery_spend":
        np.random.randint(500,10000),

        "fuel_spend":
        np.random.randint(0,5000),

        "shopping_spend":
        np.random.randint(500,15000)

    })

upi_df=pd.DataFrame(upi)

upi_df.head()
```

```python
# Save UPI Dataset
upi_df.to_csv(
    "../data/upi.csv",
    index=False)
print("upi.csv saved")
```

```python
# Generate Utility Dataset
utility=[]

for customer in customer_df["customer_id"]:

    utility.append({

        "customer_id":customer,

        "electricity_bill":
        np.random.randint(300,3000),

        "internet_bill":
        np.random.randint(300,1500),

        "delay_days":
        np.random.randint(0,30)

    })

utility_df=pd.DataFrame(utility)

utility_df.head()
```

```python
# Save Utility Dataset
utility_df.to_csv(
    "../data/utility.csv",
    index=False
)

print("utility.csv saved")
```

```python
# Generate Mobile Dataset
mobile=[]

for customer in customer_df["customer_id"]:

    mobile.append({

        "customer_id":customer,

        "recharge_frequency":
        np.random.randint(1,10),

        "monthly_data_usage":
        np.random.randint(1,100),

        "call_minutes":
        np.random.randint(50,2000)

    })

mobile_df=pd.DataFrame(mobile)

mobile_df.head()
```

```python
# Save Mobile Dataset
mobile_df.to_csv(
    "../data/mobile.csv",
    index=False)

print("mobile.csv saved")
```
