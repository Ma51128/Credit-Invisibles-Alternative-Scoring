```python
# Load Files
import pandas as pd

customer=pd.read_csv('../data/customers.csv')
upi=pd.read_csv('../data/upi.csv')
utility=pd.read_csv('../data/utility.csv')
mobile=pd.read_csv('../data/mobile.csv')
```

```python
# Merge
df=customer.merge(upi,on='customer_id')
df=df.merge(utility,on='customer_id')
df=df.merge(mobile,on='customer_id')
```

```python
# Check
df.info()
df.shape
df.isnull().sum()
df.duplicated().sum()
```

```python
# save
df.to_csv('../data/final_dataset.csv',index=False)
```
