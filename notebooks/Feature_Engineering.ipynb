```python

import pandas as pd

customer_df = pd.read_csv("../data/customers.csv")
upi_df = pd.read_csv("../data/upi.csv")
utility_df = pd.read_csv("../data/utility.csv")
mobile_df = pd.read_csv("../data/mobile.csv")
```

```python
df = customer_df.merge(
    upi_df,
    on="customer_id"
)

df = df.merge(
    utility_df,
    on="customer_id"
)

df = df.merge(
    mobile_df,
    on="customer_id"
)
```

```python
print(df.shape)
df.head()
```

```python
# Total Spend
df['total_spend'] = (
    df['grocery_spend']
    + df['fuel_spend']
    + df['shopping_spend']
)
```

```python
# savings
df['savings']=(
    df['monthly_income']
    - df['total_spend']
)
```

```python
# Income Utilization
df['utilization']=(
    df['total_spend']
    / df['monthly_income']
)
```

```python
# Payment Score
df['payment_score']=100-df['delay_days']
```

```python
# 'engagement_score
df['engagement_score']=(
    df['txn_count']
    + df['monthly_data_usage']
)
```

```python
# Create Target
df['default']=(
    (df['monthly_income']<25000)
    &
    (df['delay_days']>15)
).astype(int)
```

```python
# save
df.to_csv('../data/final_dataset.csv',index=False)
```
