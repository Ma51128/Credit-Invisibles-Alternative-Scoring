```python
import pandas as pd

df = pd.read_csv('../data/final_dataset.csv')

print(df.columns)
```

```python
# Visual 1
# Income Distribution
import seaborn as sns
sns.histplot(df['monthly_income'])
```

```python
# Visual 2
# Correlation Matrix
sns.heatmap(
    df.corr(numeric_only=True),
    annot=True
)
```

```python
print(df.columns.tolist())
'default' in df.columns
```

```python
df['default'] = (
    (df['monthly_income'] < 25000) &
    (df['delay_days'] > 15)
).astype(int)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.boxplot(
    x='default',
    y='monthly_income',
    data=df
)
plt.show()
```

```python
# Visual 4
# UPI Transactions vs Income
sns.scatterplot(
    x='txn_count',
    y='monthly_income',
    data=df
)
```
