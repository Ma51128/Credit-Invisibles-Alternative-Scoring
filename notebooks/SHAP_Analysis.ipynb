```python
import pandas as pd

df = pd.read_csv('../data/final_dataset.csv')
```

```python
X = df[
[
    'monthly_income',
    'txn_count',
    'avg_transaction',
    'delay_days',
    'monthly_data_usage',
    'recharge_frequency',
    'savings'
]
]
```

```python
import joblib

model=joblib.load(
'../models/xgb.pkl'
)
```

```python
# SHAP
import shap

explainer=shap.TreeExplainer(model)

shap_values=explainer.shap_values(X)
```

```python
# summary plot
shap.summary_plot(shap_values, X)
```

```python
shap.plots.waterfall(shap_values[0])
```

```python
import pandas as pd
import shap
import joblib

# Load dataset
df = pd.read_csv('../data/final_dataset.csv')

# Features used during training
features = [
    'monthly_income',
    'txn_count',
    'avg_transaction',
    'delay_days',
    'monthly_data_usage',
    'recharge_frequency',
    'savings'
]

X = df[features]

# Load trained model
model = joblib.load('../models/xgb.pkl')

# Create explainer
explainer = shap.Explainer(model)

# Compute SHAP values
shap_values = explainer(X)

# Summary plot
shap.plots.beeswarm(shap_values)
```

```python
shap.plots.waterfall(shap_values[0])
```
